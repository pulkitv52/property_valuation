# Phase 7: Model Evaluation

This notebook evaluates the Phase 6 RandomForest model on the held-out test split.

Scope:
- load the saved model and Phase 5 dataset
- reconstruct the test split consistently
- compute MAE, RMSE, MAPE, and R2
- generate grouped error analysis
- save metrics, predictions, and plot outputs


```python
from pathlib import Path
import json
import sys

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / 'src').exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

import geopandas as gpd

from backend.src.config import configure_logging, ensure_directories, settings
from backend.src.evaluation import build_error_analysis, evaluate_model, load_trained_model, save_evaluation_outputs

configure_logging()
ensure_directories()
PROJECT_ROOT, settings
```

```python
training_gdf = gpd.read_parquet(settings.processed_data_dir / 'model_training_dataset.parquet')
model = load_trained_model(settings.models_dir / 'valuation_model.pkl')
training_gdf.shape
```

```python
summary, predictions_df = evaluate_model(model, training_gdf)
summary
```

```python
error_analysis_df = build_error_analysis(predictions_df)
error_analysis_df.head()
```

```python
metrics_path, error_analysis_path, plot_path, predictions_path = save_evaluation_outputs(
    summary,
    predictions_df,
    error_analysis_df,
    settings.reports_dir,
    settings.processed_data_dir,
)
metrics_path, error_analysis_path, plot_path, predictions_path
```
